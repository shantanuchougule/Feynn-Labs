```python
import matplotlib.pyplot as plt
import numpy as np

# Variables
price_per_unit = 500
fixed_cost = 20000
sales = np.arange(0, 2501)  # Sales range: 0 to 2500

# Calculate revenue
revenue = (price_per_unit * sales) - fixed_cost

# Plot
plt.figure(figsize=(10, 6))
plt.plot(sales, revenue, label="Revenue = 500x - 20000", color="blue")
plt.axhline(0, color="red", linestyle="--", label="Break-even Line")
plt.title("Revenue as a Function of Sales")
plt.xlabel("Number of Subscriptions (x)")
plt.ylabel("Revenue ()")
plt.legend()
plt.grid()
plt.show()
```
